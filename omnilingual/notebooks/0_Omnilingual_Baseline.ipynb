## 0 Omnilingual Baseline

This notebook gets the baseline Word Error Rate (WER) and Character Error Rate (CER) scores for Meta's Omnilingual model.

```python
import sys
from pathlib import Path

notebook_dir = Path.cwd()
omnilingual_dir = notebook_dir.parent
sys.path.append(str(omnilingual_dir))

import pandas as pd
import torch
from tqdm import tqdm

# from omnilingual_asr.models.inference.pipeline import ASRInferencePipeline
# from omnilingual_asr.models.wav2vec2_llama.lang_ids import supported_langs
from omnilingual_asr.src.omnilingual_asr.models.inference.pipeline import ASRInferencePipeline
from omnilingual_asr.src.omnilingual_asr.models.wav2vec2_llama.lang_ids import supported_langs

def load_all_data(split="test"):
    """
    Load all test data from the data folder.
    Adjust the path to match your actual data location.
    """
    from pathlib import Path
    data_root = Path("../../data")
    all_rows = []
    for idiom_folder in data_root.iterdir():
        if not idiom_folder.is_dir():
            continue
        tsv_path = idiom_folder / f"{split}.tsv"
        if not tsv_path.exists():
            continue
        df = pd.read_csv(tsv_path, sep="\t")
        df['idiom'] = idiom_folder.name.replace("rm-", "")
        df['audio_path'] = df['path'].apply(lambda p: str(idiom_folder / "clips" / p))
        all_rows.append(df)
    return pd.concat(all_rows, ignore_index=True)
```
