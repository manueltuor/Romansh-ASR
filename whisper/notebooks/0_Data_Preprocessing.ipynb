## 0 Data Preprocessing

This notebook preprocesses the raw data, for this it will be copied into the `data/clean-data` folder and then be cleaned. Furthermore the validation set will be created for each idiom.

```python
import sys
import shutil
from pathlib import Path

notebook_dir = Path.cwd()
whisper_dir = notebook_dir.parent
sys.path.append(str(whisper_dir))

from whisper_asr.constants import CLEAN_DATA_ROOT as DATA_ROOT, RAW_DATA_ROOT, FOLDER_NAMES
from whisper_asr import preprocess_all_tsv_files, create_validation_splits, print_example
```

First we create the new directory `data/clean-data` and copy the raw data in there.

```python
DATA_ROOT.mkdir(parents=True, exist_ok=True)

shutil.copytree(RAW_DATA_ROOT, DATA_ROOT, dirs_exist_ok=True)

print("Copy complete.")
print(f"Contents of {DATA_ROOT}:")
for item in sorted(DATA_ROOT.iterdir()):
    if item.is_dir():
        print(f"    {item.name}/")
    else:
        print(f"    {item.name}")
```

The raw data still contains html tags that we need to remove. So for preprocessing we remove the html tags, and we also remove the casing and the punctuation to normalise the text.

**Note:** for correct html preprocessing we need to replace the `<br>` and `<p>` tags with spaces first, if we do not do this we will get falsely merged words.

```python
print("Cleaning HTML, punctuation and casing from all TSV files...")
preprocess_all_tsv_files(data_root=DATA_ROOT, folder_names=FOLDER_NAMES)

print("\nSentences after cleaning:")
print_example()
```

After that we subsample a validation set that constitutes of 10% of the training data for each idiom. We will save the validation split under `validation.tsv` and remove the validation samples from `train.tsv`.

```python
print("Creating validation splits...")
create_validation_splits(data_root=DATA_ROOT, folder_names=FOLDER_NAMES, val_size=0.1)
```

We can now also remove the `validated.tsv` files since these are just subsets of the `train.tsv` so they are not needed for this project.

```python
for idiom_folder in FOLDER_NAMES:
    file_path = DATA_ROOT / idiom_folder / "validated.tsv"
    if file_path.exists():
        file_path.unlink()
        print(f"  Removed {file_path}")
    else:
        print(f"  {file_path} not found")
```
