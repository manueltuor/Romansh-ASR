## 1 Data Preprocessing

This notebook preprocesses the data, furthermore the validation set will be created for each idiom.

```python
import sys
from pathlib import Path

notebook_dir = Path.cwd()
whisper_dir = notebook_dir.parent
sys.path.append(str(whisper_dir))

from whisper_asr import preprocess_all_tsv_files, create_validation_splits, print_example
from whisper_asr.constants import DATA_ROOT, FOLDER_NAMES
```

First we clean the html from the sentences in all files ending in .tsv from the `romansh-data` folder. Then we can verify that it worked with the cleaned examples from above.

**Note:** for correct html preprocessing we need to replace the `<br>` and `<p>` tags with spaces first, if we do not do this we will get falsely merged words.

```python
print("Cleaning HTML from all TSV files...")
preprocess_all_tsv_files(data_root=DATA_ROOT, folder_names=FOLDER_NAMES)

print("\nSentences after cleaning:")
print_example()
```

After that we subsample a validation set that constitutes of 10% of the training data for each idiom. We will save the validation split under `validation.tsv` and remove the validation samples from `train.tsv`.

```python
print("Creating validation splits...")
create_validation_splits(data_root=DATA_ROOT, folder_names=FOLDER_NAMES, val_size=0.1)
```
